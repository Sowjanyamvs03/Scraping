```python
url = 'https://results.eci.gov.in/AcResultGenJune2024/index.htm'
response = requests.get(url)
if response.status_code==200:
    html_content=response.text
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')  
```

```python


# Find all the items with the class "item"
items = soup.find_all('div', class_='item')

for item in items:
    state_title = item.find('h2', class_='state-title')
    if state_title:
        print('State Title:', state_title.text)

        # Find the assembly constituencies
        assembly_info = item.find('span', text='Assembly Constituencies')
        if assembly_info:
            assembly_count = assembly_info.find_next('span')
            if assembly_count:
                print('Assembly Constituencies:', assembly_count.text)
        
        # Find the partyWrap details
        party_wrap = item.find('div', class_='partyWrap')
        if party_wrap:
            print('\nParty Wrap Details:')
            print(party_wrap.text.strip())

            # Find the status of top five parties
            status_of_top_five_parties = party_wrap.find('h6', text='* Status of Top Five Parties')
            if status_of_top_five_parties:
                print('\nStatus of Top Five Parties:')
                parties_list = status_of_top_five_parties.find_next('div', class_='parties-list')
                if parties_list:
                    pr_rows = parties_list.find_all('div', class_='pr-row')
                    for pr_row in pr_rows:
                        party_name = pr_row.find_all('div')[0].text
                        leading_won = pr_row.find_all('div')[1].text
                        print(f'{party_name}: {leading_won}')
                else:
                    print('Parties list not found.')
            else:
                print('Status of Top Five Parties not found.')
        else:
            print('Party wrap details not found.')
        
        print('\n' + '-'*30 + '\n')
```

```python
from bs4 import BeautifulSoup

# Load the HTML file
with open('Andhra.html', 'r', encoding='utf-8') as file:
    content = file.read()

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(content, 'html.parser')
```

```python
# Extract the title
title = soup.title.string
print("Title of the webpage:", title)

# Extract all paragraphs
paragraphs = soup.find_all('p')
for p in paragraphs:
    print(p.text)

# Extract all headers
headers = []
for i in range(1, 7):  # Headers from h1 to h6
    headers.extend(soup.find_all(f'h{i}'))
for header in headers:
    print(header.text)

# Extract all links
links = soup.find_all('a')
for link in links:
    print(link.get('href'), link.text)

# Extract all images
images = soup.find_all('img')
for image in images:
    print(image.get('src'), image.get('alt'))

# Extract tables
tables = soup.find_all('table')
for table in tables:
    print(table.text)
```

```python
#Andhra Pradesh Results

import matplotlib.pyplot as plt

# Data
parties = ['TDP', 'JnP', 'YSRCP', 'BJP']
seats = [135, 21, 11, 8]

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(parties, seats, color=['blue', 'green', 'red', 'orange'])

# Add titles and labels
plt.title('Election Results for Andhra Pradesh - 2024')
plt.xlabel('Parties')
plt.ylabel('Number of Seats (Leading/Won)')

# Add value labels on top of the bars
for i in range(len(parties)):
    plt.text(i, seats[i] + 1, str(seats[i]), ha='center', va='bottom')

# Display the chart
plt.show()

```
